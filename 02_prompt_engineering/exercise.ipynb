```python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
client = OpenAI(api_key=os.getenv("openai_key"))

response = client.responses.create(
    model="gpt-5-nano",
    input="AI를 짧게 설명해줘",
)

print(response.output_text)
```

```python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
client = OpenAI(api_key=os.getenv("openai_key"))

# Responses API로 텍스트 생성 요청
response = client.responses.create(
    model="gpt-4.1-mini",
    input="AI 로봇에 대한 스토리를 만들어줘",          # 모델에게 보낼 프롬포트
    max_output_tokens=200,                           #  최대 출력 토큰 수
    top_p= 0.9,                                      # Top-p 샘플링(누적확률 0.9까지 후보로 답변 제한)
    temperature=0.7                                  # 생성 다양성(낮으면 안정적인 답변, 높으면 창의적인 답변)
)

print(response.output_text)                          # 응답에서 텍스트만 추출
```
