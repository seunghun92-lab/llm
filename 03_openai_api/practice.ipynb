```python
import streamlit as st
from openai import OpenAI
from dotenv import load_dotenv
import os

# 1) .env ë¡œë“œ
load_dotenv()
api_key = os.getenv("openai_key")

# 2) OpenAI í´ë¼ì´ì–¸íŠ¸
client = OpenAI(api_key=api_key)

st.set_page_config(page_title="My Chatbot", page_icon="ğŸ¤–")
st.title("ğŸ¤– Streamlit ì±—ë´‡")

# 3) ëŒ€í™” ê¸°ë¡ ì €ì¥ì†Œ ë§Œë“¤ê¸° (ì²˜ìŒ ì‹¤í–‰ë  ë•Œë§Œ)
if "messages" not in st.session_state:
    st.session_state.messages = [
        {"role": "system", "content": "ë„ˆëŠ” ì¹œì ˆí•œ í•œêµ­ì–´ ì±—ë´‡ì´ì•¼. ì§§ê³  ëª…í™•í•˜ê²Œ ë‹µí•´ì¤˜."}
    ]

# 4) í™”ë©´ì— ì§€ê¸ˆê¹Œì§€ ëŒ€í™” ë³´ì—¬ì£¼ê¸° (systemì€ ìˆ¨ê¹€)
for msg in st.session_state.messages:
    if msg["role"] == "system":
        continue
    with st.chat_message(msg["role"]):
        st.write(msg["content"])

# 5) ì…ë ¥ì°½
user_input = st.chat_input("ë©”ì‹œì§€ë¥¼ ì…ë ¥í•˜ì„¸ìš”...")

if user_input:
    # (1) ì‚¬ìš©ì ë©”ì‹œì§€ ì €ì¥ + í™”ë©´ í‘œì‹œ
    st.session_state.messages.append({"role": "user", "content": user_input})
    with st.chat_message("user"):
        st.write(user_input)

    # (2) ëª¨ë¸ í˜¸ì¶œ
    with st.chat_message("assistant"):
        with st.spinner("ìƒê° ì¤‘..."):
            response = client.chat.completions.create(
                model="gpt-4.1-mini",
                messages=st.session_state.messages,
                temperature=0.7
            )
            answer = response.choices[0].message.content

            st.write(answer)

    # (3) ì–´ì‹œìŠ¤í„´íŠ¸ ë‹µ ì €ì¥
    st.session_state.messages.append({"role": "assistant", "content": answer})
```
